```python
import sympy as smp
import matplotlib.pyplot as plt
```

```python
f, fj1m, fj1p, fj2m, fj2p  = smp.symbols(r'f f_{j-1} f_{j+1} f_{j-2} f_{j+2}')
df, d2f, d3f, d4f, d5f, d6f = smp.symbols(r'f^{(1)} f^{(2)} f^{(3)} f^{(4)} f^{(5)} f^{(6)}')
x, h  = smp.symbols(r'x \Delta')
```

# Lecture Example

```python
taylor = f + df*h + d2f/2*h**2 + d3f/6*h**3 + d4f/24*h**4 
taylor 
```

```python
f1p = taylor.subs(h,h) - fj1p
f1m = taylor.subs(h,-h) - fj1m
f2p = taylor.subs(h,h*2) - fj2p
f2m = taylor.subs(h,h*2) - fj2m
```

```python
smp.solve(f1p - f1m,df)[0].expand()
```

```python
smp.solve(f1p + f1m,d2f)[0].expand()
```

```python
taylor = f + df*h + d2f/2*h**2 + d3f/6*h**3 + d4f/24*h**4 + d5f/120*h**5  + d6f/720*h**6
taylor 
```

```python
f1p = taylor.subs(h,h) - fj1p
f1m = taylor.subs(h,-h) - fj1m
f2p = taylor.subs(h,2*h) - fj2p
f2m = taylor.subs(h,-2*h) - fj2m
```

```python
f1_sum = f1p + f1m
f1_sum
```

```python
d2f_plug = smp.solve(f1_sum, d2f)[0] 
d2f_plug
```

```python
f2_sum = f2p + f2m
f2_sum = f2_sum.subs(d2f,d2f_plug)
f2_sum
```

```python
ans = smp.solve(f2_sum, d4f)[0] 
ans
```
