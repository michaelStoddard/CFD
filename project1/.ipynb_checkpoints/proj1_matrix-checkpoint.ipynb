# Domain and driving funciton

```python
# boundary conditions
def BC_west(y):
    return 0.5 - 0.5 * np.cos(2 * np.pi * y)

def BC_east(y):
    return 0

def BC_south(x):
    return -0.3 

def BC_north(x):
    return 0.5 + 0.5 * np.sin(4*np.pi * x - 0.5*np.pi)


```

```python
def kappa(x,y):
    return 0.02*np.exp(-((x-0.7)**2/0.09 + (y-0.6)**2/0.25))
```

## Building Laplace Operator

$$\boxed{f_{BC} = \frac{-2}{3\Delta}f'_{BC} - \frac{2}{3\Delta^2}f_{j+1} + \frac{2}{3\Delta^2}f_{j+2}}$$
Where $f'_{BC}$ is given as the boundary condition

Second derivative, Second order accurate `L2`, with main diagonal, central stencil, `m2d2:`
$$\frac{1}{\Delta^2}\begin{bmatrix}1&-2&1\end{bmatrix}\begin{bmatrix}f_{j-1}\\f_j\\ f_{j+1}\end{bmatrix}$$

$$\boxed{f_{BC} = \frac{-2\Delta}{3}f'_{BC} + \frac{4}{3}f_{j+1} - \frac{1}{3}f_{j+2}}$$
and
$$\boxed{f''_{BC} = \frac{-2}{3\Delta}f'_{BC} - \frac{2}{3\Delta^2}f_{j+1} + \frac{2}{3\Delta^2}f_{j+2}}$$
Where $f'_{BC}$ is given as the boundary condition

```python
north = np.arange(0,Ny,1)[1:-1]
west = np.arange(0,N,Ny)[1:-1]
south = np.arange(N-Ny,N,1)[1:-1]
east = np.arange(Ny-1,N,Ny)[1:-1]

four = np.array([0,Ny-1,-Ny,-1])
nw = 0
sw = -Ny
se = -1
ne = Ny-1


clear = np.zeros(N)
```

## My stupid method

```python
L2 = np.zeros((Nx*Ny,Nx*Ny))

# First define the matrix with the central stenicl along the entire diagonal
m2d2 = np.array([1/dy**2, 1/dx**2,
                 -2*(1/dx**2 + dy**2),
                 1/dx**2, 1/dy**2])

L2 = spMtrx(m2d2,
              [-Ny,-1, 0, 1, Ny],
              shape=(N,N)).toarray()

L2[north] = clear
L2[west] = clear
L2[south] = clear
L2[east] = clear

L2[four] = clear


########### North ###############

# Ly: Dir ----------------------------
# L2[north,north]      = BC_north(x[1:-1])/dy**2 
L2[north,north+Ny]   = -2/dy**2 
L2[north,north+2*Ny] = 1/dy**2 

## Lx: Cen ----------------------------
L2[north,north]  += -2/dx**2 
L2[north,north+1] = 1/dx**2
L2[north,north-1] = 1/dx**2


############ West ###############

## Lx: Dir ----------------------------
# L2[west,west]   = BC_west(y[1:-1])/dx**2 
L2[west,west+1] = - 2/dx**2
L2[west,west+2] =  1/dx**2

# Ly: Cen ----------------------------
L2[west,west]   += -2/dy**2 
L2[west,west+Ny] = 1/dy**2
L2[west,west-Ny] = 1/dy**2



############ South ###############

# Ly: Neu ----------------------------
# L2[south,south] = -2/(3*dy)*BC_south(x[1:-1])
L2[south,south-Ny] = -2/(3*dy**2) 
L2[south,south-2*Ny] = +2/(3*dy**2) 

## Lx: Cen ----------------------------
L2[south,south]  += -2/dx**2
L2[south,south+1] = 1/dx**2
L2[south,south-1] = 1/dx**2



############ East ###############

## Lx: Neu ----------------------------
# L2[east,east]   = -2/(3*dx)*BC_east(y[1:-1])
L2[east,east-1] = -2/(3*dx**2) 
L2[east,east-2] = +2/(3*dx**2) 

# Ly: Cen ----------------------------
L2[east,east]   += -2/dy**2
L2[east,east+Ny] = 1/dy**2
L2[east,east-Ny] = 1/dy**2



############ Four ###############
## ----------------------------

### NW corner (west dirichlet + north dirichlet)
## dX ----------------------------
# L2[nw,nw]   = BC_north(x[1:-1]+1)/dx**2
# L2[nw,nw+1] = BC_north(x[1:-1]+1)/dx**2
# L2[nw,nw+2] = BC_north(x[1:-1]-1)/dx**2

# L2[nw,nw]    = - 2*(1/dx**2 + 1/dy**2) 
# L2[nw,nw+1]  = 1/dx**2
# L2[nw,nw+Ny] = 1/dy**2



# # southwest corner (west dirichlet + south neumann)
# L2[sw,sw]    = - 2*(1/dx**2 + 1/dy**2)
# L2[sw,sw+1]  = 1/dx**2
# L2[sw,sw-Ny] = 1/dy**2

# # southeast corner (east neumann + south neumann)
# L2[se,se] = - 2*(1/dx**2 + 1/dy**2)
# L2[se,se-1] = 1/dx**2
# L2[se,se-Ny] = 1/dy**2


# # northeast corner (east neumann + north dirichlet)
# L2[ne,ne]    = - 2*(1/dx**2 + 1/dy**2)
# L2[ne,ne-1]  = 1/dx**2
# L2[ne,ne+Ny] = 1/dy**2
```

* Boundary Conditions as stencils 
    * Second derivative, Neuman: $f''_{BC} = \frac{-2}{3\Delta}f'_{BC} - \frac{2}{3\Delta^2}f_{j+1} + \frac{2}{3\Delta^2}f_{j+2}$
    * Second derivative, Dirichlet: $f''_{BC} = \frac{f_{BC}}{\Delta^2} - \frac{2}{\Delta^2}f_{j+1} + \frac{1}{\Delta^2}f_{j+2}$
    * Function, Neuman: $f_{BC} = \frac{-2\Delta}{3}f'_{BC} + \frac{4}{3}f_{j+1} - \frac{1}{3}f_{j+2}$
    


```python
def laplace(u,alpha=0.005):
    v = L2@u
    
    v[north] += BC_north(x[1:-1])/dy**2 
    v[west]  += BC_west(y[1:-1])/dx**2 
    v[south] += -2/(3*dy)*BC_south(x[1:-1])
    v[east]  += -2/(3*dx)*BC_east(y[1:-1])
    
    v[mw]  = BC_north(x=0)#/dy**2 - 2/dy**2*BC_west(y[1]) + 1/dy**2*BC_west(y[2])
    # v[mw] += BC_west(y[0])/dx**2 - 2/dx**2*BC_north(x[1]) + 1/dx**2*BC_north(x[2])
    
    v[sw]  = BC_west(y=0)
    
    v[se]  += -2/(3*dy)*BC_south(x=1)
    
    v[ne]  = BC_north(x=1)
    
    return v*alpha
```

```python

def enforc_BC(v,alpha=0.005):
    # v = U_BC@u
    
    v[north] = BC_north(x)
    
    v[west] = BC_west(y)

    v[east] = BC_east(y)  
    
    v[south] = -BC_south(x) 

    
    v[sw] = BC_west(y=0) # - BC_south(x=0)
    
    v[se] = -BC_south(x=1) #+ BC_east(y=0) 
    
    v[nw] = BC_west(y=1) # + BC_north(x=0)/dy**2
    
    v[ne] = BC_north(x=1) # + BC_east(y=1)         
    
    return v
```

## Nair's Method

```python
L2 = np.zeros((Nx*Ny,Nx*Ny))

# First define the matrix with the central stenicl along the entire diagonal
m2d2 = np.array([1/dy**2,1/dx**2,-2*(1/dx**2 + dy**2),1/dx**2,1/dy**2])

L2 = spMtrx(m2d2,
              [-Ny,-1, 0, 1, Ny],
              shape=(N,N)).toarray()

L2[west,west] = -2*(1/dx**2 + 1/dy**2)
L2[west,west+1] = 1/dx**2
L2[west,west+Ny] = 1/dy**2
L2[west,west-Ny] = 1/dy**2

L2[north,north] = -2*(1/dx**2 + 1/dy**2)
L2[north,north+Ny] = 1/dx**2 #This may need to be flipped: -Ny
L2[north,north+1] = 1/dx**2
L2[north,north-1] = 1/dx**2

L2[east,east] = -1/dx**2 - 2/dy**2
L2[east,east-1] = 1/dx**2
L2[east,east+Ny] = 1/dy**2
L2[east,east-Ny] = 1/dy**2

L2[south,south] = -2/dx**2 - 1/dy**2
L2[south,south-Ny] = 1/dy**2 #This may need to be flipped: +Ny
L2[south,south+1] = 1/dx**2
L2[south,south-1] = 1/dx**2

# corners
# southwest corner (west dirichlet + south neumann)
L2[sw,sw]    = -2/dx**2 - 1/dy**2
L2[sw,sw+1]  = 1/dx**2
L2[sw,sw-Ny] = 1/dy**2

# southeast corner (east neumann + south neumann)
L2[se,se] = -1/dx**2 -1/dy**2
L2[se,se-1] = 1/dx**2
L2[se,se-Ny] = 1/dy**2

# northwest corner (west dirichlet + north dirichlet)
L2[nw,nw]    = -2/dx**2 - 2/dy**2 
L2[nw,nw+1]  = 1/dx**2
L2[nw,nw+Ny] = 1/dy**2

# northeast corner (east neumann + north dirichlet)
L2[ne,ne]    = - 1/dx**2 - 2/dy**2
L2[ne,ne-1]  = 1/dx**2
L2[ne,ne+Ny] = 1/dy**2
```

```python
def LHS(u,alpha=0.005):
    v = L2@u
    return  -v*alpha

def BC_west(y):
    return 0.5 - 0.5 * np.cos(2 * np.pi * y)

def BC_east(y):
    return np.zeros(len(y))

def BC_south(x):
    return -0.3 * np.ones(len(x))

def BC_north(x):
    return 0.5 + 0.5 * np.sin(4*np.pi * x - 0.5*np.pi)

def RHS(alpha=0.005):
    # v = U_BC@u
    v = np.zeros(K.shape)
    
    v[north] = BC_north(x[1:-1])/dy**2
    
    v[west] = BC_west(y[1:-1])/dx**2

    v[east] = BC_east(y[1:-1])/dx  
    
    v[south] = -BC_south(x[1:-1])/dy  

    
    v[sw] = BC_west(y=0)/dx**2 - BC_south(x=0)/dy
    
    v[se] = BC_east(y=0)/dx - BC_south(x=1)/dy 
    
    v[nw] = BC_west(y=1)/dx**2 + BC_north(x=0)/dy**2
    
    v[ne] = BC_east(y=1)/dx + BC_north(x=1)/dy**2         
    
    return K + alpha*v


def conjGrad(LHS, RHS, x0=None, tol=1e-5, maxit=1000):
    if x0 is None:
        u = np.zeros(N)
    else:
        u = x0
    r = RHS - LHS(u)
    p = r.copy()
    res0 = np.linalg.norm(r)
    resvec = np.zeros(maxit)

    for iter in range(1, maxit + 1):
        LHS_p = LHS(p)
        alpha = (r @ r) / (p @ LHS_p)

        u = u + alpha * p
        r_new = r - alpha * LHS_p

        resvec[iter - 1] = np.linalg.norm(r_new) / res0

        if resvec[iter - 1] < tol:
            resvec = resvec[:iter]
            return u, iter, resvec

        beta = (r_new @ r_new) / (r @ r)
        p = r_new + beta * p
        r = r_new

    resvec = resvec[:iter]
    return u, iter, resvec
```

```python
[U, iter, resvec] = conjGrad(laplace,-K)
print(f'CG converged in {iter} iterations with final residual {resvec[-1]:.2e}')
```
