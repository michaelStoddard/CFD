```python
import sympy as smp
```

Defining the symbolic variables, that will be used to construct the Taylor series

```python
f, fj1m, fj1p, fj2m, fj2p  = smp.symbols(r'f f_{j-1} f_{j+1} f_{j-2} f_{j+2}')
df, d2f, d3f, d4f, d5f, d6f = smp.symbols(r'f^{(1)} f^{(2)} f^{(3)} f^{(4)} f^{(5)} f^{(6)}')
x, h  = smp.symbols(r'x \Delta')
```

# Lecture Example

```python
taylor = f + df*h + d2f/2*h**2 + d3f/6*h**3 + d4f/24*h**4 
taylor 
```

```python
f1p = taylor.subs(h,h) - fj1p
f1m = taylor.subs(h,-h) - fj1m
f2p = taylor.subs(h,h*2) - fj2p
f2m = taylor.subs(h,h*2) - fj2m
```

```python
smp.solve(f1p - f1m,df)[0].expand()
```

```python
smp.solve(f1p + f1m,d2f)[0].expand()
```

# Problem 1

Defining the symbolic Taylor series, up to the $6^{th}$ derivative

```python
taylor = f + df*h + d2f/2*h**2 + d3f/6*h**3 + d4f/24*h**4 + d5f/120*h**5  + d6f/720*h**6
taylor 
```

Creating the needed functions by replacing the $\Delta$ in the above equation. Note, to get the $f_{j+n}$ in the symbolic solution it has to be subtracted over to the RHS of the Taylor series. For the following, $f\equiv f_j$

* `f1p` $\to0= -f_{j+1} + \Delta f' + \frac{\Delta^2f''}{2!}+...+\mathcal{O}^6$
* `f1m` $\to0= -f_{j-1} - \Delta f' + \frac{(-\Delta)^2f''}{2!}+...+\mathcal{O}^6$
* `f2p` $\to0= -f_{j+2} + \Delta f' + \frac{(2\Delta)^2f''}{2!}+...+\mathcal{O}^6$
* `f2m` $\to0= -f_{j-2} - \Delta f' + \frac{(-2\Delta)^2f''}{2!}+...+\mathcal{O}^6$

```python
f1p = taylor.subs(h,h) - fj1p
f1m = taylor.subs(h,-h) - fj1m
f2p = taylor.subs(h,2*h) - fj2p
f2m = taylor.subs(h,-2*h) - fj2m
```

Getting a series of even powers by 

`f1_sum` $\to f_{j+1}+f_{j-1}$

```python
f1_sum = f1p + f1m
f1_sum
```

Solving the sum $f_{j+1}+f_{j-1}$ for the second derivative

`d2f_plug` $\to f''$

```python
d2f_plug = smp.solve(f1_sum, d2f)[0] 
d2f_plug
```

Similarly, doing the sum, $f_{j+2} + f_{j-2}\to$ `f2_sum`, and plugging in the second derivative, $f''\to$ `d2f_plug`, which was found in the above step

```python
f2_sum = f2p + f2m
f2_sum = f2_sum.subs(d2f,d2f_plug)
f2_sum
```

Finally, the answer is given by solving the above, `f2_sum`, for the fourth derivative, `d4f` $\to f^{(4)}$

```python
ans = smp.solve(f2_sum, d4f)[0] 
ans
```

$$\boxed{f^{(4)}_j = \frac{f_{j-2} - 4f_{j-1} +6f_j - 4f_{j+1}+f_{j+2}}{\Delta^4}+\mathcal{O}(6)}$$

# Problem 2

Creating the needed functions by replacing the $\Delta$ in the general Taylor variable, `taylor`, that was created in problem 1. In this case, the $\Delta$ is substituted as follows

* `f1p` $\to0= -f_{j+1} + (\Delta/2) f' + \frac{(\Delta/2)^2f''}{2!}+...+\mathcal{O}^6$
* `f1m` $\to0= -f_{j-1} + (-\Delta/2) f' + \frac{(-\Delta/2)^2f''}{2!}+...+\mathcal{O}^6$
* `f2p` $\to0= -f_{j+2} + (3\Delta/2) f' + \frac{(3\Delta/2)^2f''}{2!}+...+\mathcal{O}^6$
* `f2m` $\to0= -f_{j-2} + (-3\Delta/2) f' + \frac{(-3\Delta/2)^2f''}{2!}+...+\mathcal{O}^6$

```python
f1p = taylor.subs(h,h/2) - fj1p
f1m = taylor.subs(h,-h/2) - fj1m
f2p = taylor.subs(h,3*h/2) - fj2p
f2m = taylor.subs(h,-3*h/2) - fj2m
```

Getting a series of odd powers by 

`f1_dif:` $f_{j+1}-f_{j-1}$

```python
f1_dif = f1p - f1m
f1_dif
```

Similarly, getting another series of odd powers by 

`f2_sum:` $f_{j+2}+f_{j-2}$

```python
f2_dif = f2p - f2m
f2_dif
```

Solving `f2_dif` for the the third derivative, $f'''\to$ `d3f_plug`

```python
d3f_plug = smp.solve(f2_dif,d3f)[0]
d3f_plug
```

Plugging the result into `f1_diff`

```python
f1_dif = f1_dif.subs(d3f,d3f_plug)
f1_dif
```

Then the answer is given by solving for the first derivative, $f'\to$ `df`

```python
ans = smp.solve(f1_dif,df)[0].expand()
ans
```

$$\boxed{f^{(1)}_j = \frac{f_{j-2} - 27f_{j-1} + 27f_{j+1}-f_{j+2}}{24\Delta}+\mathcal{O}(4)}$$

# Problem 3

Creating the needed functions by replacing the $\Delta$ in the general Taylor variable, `taylor`, that was created in problem 1. In this case, the $\Delta$ is substituted as follows

* `f1p`$\to0= -f_{j+1} + \Delta f' + \frac{(\Delta)^2f''}{2!}+...+\mathcal{O}^6$
* `f2p`$\to0= -f_{j+2} + 2\Delta f' + \frac{(2\Delta)^2f''}{2!}+...+\mathcal{O}^6$

```python
f1p = taylor.subs(h,h) - fj1p
f2p = taylor.subs(h,2*h) - fj2p
```

```python
f1p
```

```python
f2p
```

In observing the above two Taylor series, we can cancel out the first derivative by subtracting $2*f_{j+1}$ from $f_{j+2}$

```python
f_dif = f2p - 2*f1p
f_dif
```

This result can be solved for the second derivative

`d2f_plug` $\to f''$

```python
d2f_plug = smp.solve(f_dif,d2f)[0]
d2f_plug
```

Now solving for the first derivative by plugging this result into the sum $2f_{j+1} + f_{j+2}$

```python
f_sum = 2*f1p + f2p
f_sum = f_sum.subs(d2f,d2f_plug)

ans = smp.solve(f_sum,df)[0].expand()
ans
```

$$\boxed{f^{(1)}_j = \frac{-3f_{j} + 4f_{j+1}-f_{j+2}}{2\Delta}+\mathcal{O}(2)}$$
